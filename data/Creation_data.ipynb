```python
import pandas as pd
import unicodedata

# 1. Toujours notre outil pour nettoyer les noms (minuscules, sans accents)
def normaliser_nom(nom):
    if not isinstance(nom, str):
        return ""
    nom_sans_accents = ''.join(c for c in unicodedata.normalize('NFD', nom)
                               if unicodedata.category(c) != 'Mn')
    return nom_sans_accents.lower().strip()

# 2. Votre liste brute copiée-collée
texte_fifa = """France - 1
Spain - 2
Argentina - 3
England - 4
Portugal - 5
Brazil - 6
Morocco - 7
Netherlands - 8
Belgium - 9
Germany - 10
Croatia - 11
Colombia - 13
Senegal - 14
Mexico - 15
USA - 16
Uruguay - 17
Japan - 18
Switzerland - 19
Iran - 21
Turkey - 22
Austria - 23
Ecuador - 24
South Korea - 25
Australia - 27
Algeria - 28
Egypt - 29
Canada - 30
Norway - 31
Panama - 33
Ivory Coast - 34
Sweden - 38
Paraguay - 40
Czechia - 41
Scotland - 43
DR Congo - 45
Tunisia - 46
Uzbekistan - 50
Qatar - 55
Iraq - 57
South Africa - 60
Saudi Arabia - 61
Jordan - 63
Bosnia and Herzegovina - 64
Cape Verde - 68
Ghana - 73
Haiti - 82
Curaçao - 83
New Zealand - 85"""

# 3. Petit dictionnaire de traduction (Anglais -> Français) 
# Ajoutez-en d'autres ici si vous voyez que certaines équipes ne se mettent pas à jour
traductions = {
    "spain": "espagne",
    "brazil": "bresil",
    "morocco": "maroc",
    "netherlands": "pays-bas",
    "germany": "allemagne",
    "croatia": "croatie",
    "colombia": "colombie",
    "mexico": "mexique",
    "switzerland": "suisse",
    "turkey": "turquie",
    "austria": "autriche",
    "ecuador": "equateur",
    "south korea": "coree du sud",
    "algeria": "algerie",
    "egypt": "egypte",
    "ivory coast": "cote d'ivoire",
    "sweden": "suede",
    "czechia": "republique tcheque",
    "scotland": "ecosse",
    "dr congo": "rd congo",
    "south africa": "afrique du sud",
    "saudi arabia": "arabie saoudite",
    "cape verde": "cap-vert",
    "new zealand": "nouvelle-zelande"
}

# --- CRÉATION DU DICTIONNAIRE DE CLASSEMENT ---
dico_classement = {}

# On découpe le grand texte ligne par ligne
for ligne in texte_fifa.strip().split('\n'):
    if ' - ' in ligne:
        equipe_anglais, rang = ligne.split(' - ')
        
        # On normalise le nom anglais
        nom_normalise = normaliser_nom(equipe_anglais)
        
        # Si on a une traduction prévue, on l'utilise, sinon on garde le nom tel quel
        nom_final = traductions.get(nom_normalise, nom_normalise)
        
        # On enregistre dans notre dictionnaire (ex: {'bresil': 6})
        dico_classement[nom_final] = int(rang)


# --- MISE À JOUR DU DATAFRAME ---

print("🚀 Mise à jour du classement FIFA en cours...\n")

equipes_non_trouvees = []

for index, row in df.iterrows():
    nom_df_nettoye = normaliser_nom(row['Team_name'])
    
    # On cherche l'équipe dans notre dictionnaire fraîchement créé
    if nom_df_nettoye in dico_classement:
        df.at[index, 'Fifa_ranking'] = dico_classement[nom_df_nettoye]
    else:
        equipes_non_trouvees.append(row['Team_name'])

# --- RÉSULTATS ---
print("✅ Classements mis à jour avec succès !")
print("\n📊 Aperçu du DataFrame :")
print(df[['Team_name', 'Fifa_ranking']])

if equipes_non_trouvees:
    print("\n⚠️ Attention, ces équipes n'ont pas trouvé leur classement (vérifiez l'orthographe ou ajoutez-les dans le dictionnaire 'traductions') :")
    for eq in equipes_non_trouvees:
        print(f"- {eq}")
```

```python
"""
ÉTAPE 1 — Scraping Flashscore (Selenium + BeautifulSoup)
Récupère les 80 derniers matchs pour chaque équipe.
Sortie : data/matches_raw.csv

Prérequis :
    pip install selenium webdriver-manager beautifulsoup4 pandas
"""

import re
import time
import pandas as pd
from pathlib import Path
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from webdriver_manager.chrome import ChromeDriverManager

# ── Config ────────────────────────────────────────────────────────────────────
INPUT_CSV        = Path("stats_equipe.csv")
OUTPUT_CSV       = Path("data/matches_raw.csv")
N_MATCHES        = 80     # matchs cibles par équipe
MAX_CLICS        = 1      # ~13 matchs/clic → 7 clics ≈ 90 matchs chargés
HEADLESS         = False  # True pour cacher Chrome
DELAY_PAGE       = 3      # secondes après chargement page
DELAY_CLIC       = 2      # secondes après chaque "Montrer plus"
DELAY_BTW_TEAMS  = 2      # secondes de pause entre équipes

# ── Driver ────────────────────────────────────────────────────────────────────

def make_driver() -> webdriver.Chrome:
    options = Options()
    if HEADLESS:
        options.add_argument("--headless=new")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--disable-blink-features=AutomationControlled")
    options.add_argument(
        "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/124.0.0.0 Safari/537.36"
    )
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    return webdriver.Chrome(
        service=Service(ChromeDriverManager().install()),
        options=options,
    )

# ── Helpers ───────────────────────────────────────────────────────────────────

def parse_score(text: str):
    """'2' ou '2 ' → int, ou None si non numérique."""
    m = re.search(r"\d+", str(text))
    return int(m.group()) if m else None

def result_label(goals_for, goals_against) -> str | None:
    """Retourne 'W', 'D' ou 'L' du point de vue de l'équipe cible."""
    if goals_for is None or goals_against is None:
        return None
    if goals_for > goals_against:
        return "W"
    if goals_for == goals_against:
        return "D"
    return "L"

def accept_cookies(driver):
    """Ferme la bannière cookies si elle apparaît."""
    selectors = [
        "#onetrust-accept-btn-handler",
        "button#accept-all",
        "[data-testid='cookie-policy-dialog-accept-btn']",
        "button.fc-button-label",          # Flashscore FR variante
    ]
    for sel in selectors:
        try:
            btn = WebDriverWait(driver, 3).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, sel))
            )
            btn.click()
            time.sleep(1)
            return
        except Exception:
            continue

def click_show_more(driver, max_clics: int):
    """Clique sur 'Montrer plus' / 'Show more' jusqu'à max_clics fois."""
    xpaths = [
        '//button[contains(., "Montrer plus")]',
        '//button[contains(., "Show more")]',
        '//a[contains(@class,"event__more")]',
        '//div[contains(@class,"showMore")]',
    ]
    clicked = 0
    for _ in range(max_clics):
        found = False
        for xp in xpaths:
            try:
                btn = WebDriverWait(driver, 3).until(
                    EC.presence_of_element_located((By.XPATH, xp))
                )
                driver.execute_script(
                    "arguments[0].scrollIntoView({block:'center'});", btn
                )
                time.sleep(0.5)
                driver.execute_script("arguments[0].click();", btn)
                clicked += 1
                time.sleep(DELAY_CLIC)
                found = True
                break
            except Exception:
                continue
        if not found:
            break   # plus de bouton dispo
    return clicked

# ── Scraper par équipe ────────────────────────────────────────────────────────

def scrape_team(driver, team_name: str, url: str) -> list[dict]:
    """
    Scrape jusqu'à N_MATCHES résultats pour une équipe.
    Retourne une liste de dicts avec toutes les colonnes du pipeline.
    """
    print(f"  → {team_name}")
    rows = []

    try:
        driver.get(url)
        time.sleep(DELAY_PAGE)

        accept_cookies(driver)
        n_clics = click_show_more(driver, MAX_CLICS)
        print(f"     {n_clics} clics 'Montrer plus'")

        soup = BeautifulSoup(driver.page_source, "html.parser")
        matchs = soup.find_all(
            "div", class_=lambda x: x and "event__match" in x
        )
        print(f"     {len(matchs)} éléments match trouvés")

        for match in matchs[:N_MATCHES]:

            # ── Résultat via badge de forme ──────────────────────────────
            # (badge présent sur la page résultats Flashscore)
            badge = match.find(
                "button",
                class_=lambda x: x and "formIcon__lastMatches" in x,
            )
            badge_result = badge.text.strip() if badge else None
            # V=win, N=draw, D=loss (labels FR de Flashscore)
            result_from_badge = {"V": "W", "N": "D", "D": "L"}.get(
                badge_result, None
            )

            # ── Équipes & score ──────────────────────────────────────────
            try:
                elem_dom = match.find(
                    class_=lambda x: x and "event__homeParticipant" in x
                )
                elem_ext = match.find(
                    class_=lambda x: x and "event__awayParticipant" in x
                )
                elem_score_dom = match.find(
                    class_=lambda x: x and "event__score--home" in x
                )
                elem_score_ext = match.find(
                    class_=lambda x: x and "event__score--away" in x
                )

                if not all([elem_dom, elem_ext, elem_score_dom, elem_score_ext]):
                    continue

                home_team = elem_dom.text.strip()
                away_team = elem_ext.text.strip()
                score_dom = parse_score(elem_score_dom.text)
                score_ext = parse_score(elem_score_ext.text)

                if score_dom is None or score_ext is None:
                    continue

                # L'équipe cible est-elle à domicile ?
                is_home = team_name.lower() in home_team.lower()

                if is_home:
                    goals_for     = score_dom
                    goals_against = score_ext
                else:
                    goals_for     = score_ext
                    goals_against = score_dom

                # Résultat final : badge en priorité, score en fallback
                result = result_from_badge or result_label(goals_for, goals_against)

            except Exception:
                continue

            # ── Date ─────────────────────────────────────────────────────
            date_el = match.find(
                class_=lambda x: x and "event__time" in x
            )
            date_str = date_el.text.strip() if date_el else ""

            # ── Compétition ───────────────────────────────────────────────
            # On remonte au parent pour chercher le header de compétition
            comp = "Unknown"
            try:
                parent = match.find_parent(
                    "div", class_=lambda x: x and "event__round" in x
                )
                if parent:
                    comp = parent.text.strip()
            except Exception:
                pass

            rows.append({
                "team":          team_name,
                "date":          date_str,
                "home_team":     home_team,
                "away_team":     away_team,
                "home_goals":    score_dom,
                "away_goals":    score_ext,
                "is_home":       is_home,
                "goals_for":     goals_for,
                "goals_against": goals_against,
                "result":        result,
                "competition":   comp,
            })

    except Exception as e:
        print(f"  ⚠ Erreur {team_name} : {e}")

    return rows

# ── Pipeline principal ────────────────────────────────────────────────────────

def run_scraper(df: pd.DataFrame | None = None) -> pd.DataFrame:
    """
    Lance le scraping pour toutes les équipes.

    Args:
        df : DataFrame avec colonnes Team_name & links_flashscore.
             Si None, charge INPUT_CSV automatiquement.

    Returns:
        DataFrame des matchs bruts (sauvegardé dans OUTPUT_CSV).
    """
    if df is None:
        df = pd.read_csv(INPUT_CSV)

    all_matches: list[dict] = []
    driver = make_driver()

    try:
        for idx, row in df.iterrows():
            team = row["Team_name"]
            url  = row.get("links_flashscore", "")

            if pd.isna(url) or not str(url).startswith("http"):
                print(f"  ⚠ URL manquante pour {team}, skip")
                continue

            matches = scrape_team(driver, team, str(url))
            all_matches.extend(matches)
            print(f"     ✓ {len(matches)} matchs récupérés\n")

            time.sleep(DELAY_BTW_TEAMS)

    finally:
        driver.quit()   # toujours fermer même si erreur

    results_df = pd.DataFrame(all_matches)
    OUTPUT_CSV.parent.mkdir(parents=True, exist_ok=True)
    results_df.to_csv(OUTPUT_CSV, index=False)
    print(f"\n✅ Scraping terminé → {len(results_df)} matchs au total")
    print(f"   Sauvegardé : {OUTPUT_CSV}")
    return results_df


# ── Point d'entrée ────────────────────────────────────────────────────────────

if __name__ == "__main__":
    run_scraper()
```

```python
"""
ÉTAPE 1 — Scraping Flashscore (Selenium + BeautifulSoup)
Récupère les 80 derniers matchs pour chaque équipe.
Sortie : data/matches_raw.csv

Prérequis :
    pip install selenium webdriver-manager beautifulsoup4 pandas
"""

import re
import time
import pandas as pd
from pathlib import Path
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from webdriver_manager.chrome import ChromeDriverManager

# ── Config ────────────────────────────────────────────────────────────────────
INPUT_CSV        = Path("stats_equipe.csv")
OUTPUT_CSV       = Path("matches_raw.csv")
N_MATCHES        = 80     # matchs cibles par équipe
MAX_CLICS        = 1      # ~13 matchs/clic → 7 clics ≈ 90 matchs chargés
HEADLESS         = False  # True pour cacher Chrome
DELAY_PAGE       = 3      # secondes après chargement page
DELAY_CLIC       = 2      # secondes après chaque "Montrer plus"
DELAY_BTW_TEAMS  = 2      # secondes de pause entre équipes

# ── Driver ────────────────────────────────────────────────────────────────────

def make_driver() -> webdriver.Chrome:
    options = Options()
    if HEADLESS:
        options.add_argument("--headless=new")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--disable-blink-features=AutomationControlled")
    options.add_argument(
        "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/124.0.0.0 Safari/537.36"
    )
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    return webdriver.Chrome(
        service=Service(ChromeDriverManager().install()),
        options=options,
    )

# ── Helpers ───────────────────────────────────────────────────────────────────

def parse_score(text: str):
    """'2' ou '2 ' → int, ou None si non numérique."""
    m = re.search(r"\d+", str(text))
    return int(m.group()) if m else None

def result_label(goals_for, goals_against) -> str | None:
    """Retourne 'W', 'D' ou 'L' du point de vue de l'équipe cible."""
    if goals_for is None or goals_against is None:
        return None
    if goals_for > goals_against:
        return "W"
    if goals_for == goals_against:
        return "D"
    return "L"

def accept_cookies(driver):
    """Ferme la bannière cookies si elle apparaît."""
    selectors = [
        "#onetrust-accept-btn-handler",
        "button#accept-all",
        "[data-testid='cookie-policy-dialog-accept-btn']",
        "button.fc-button-label",          # Flashscore FR variante
    ]
    for sel in selectors:
        try:
            btn = WebDriverWait(driver, 3).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, sel))
            )
            btn.click()
            time.sleep(1)
            return
        except Exception:
            continue

def click_show_more(driver, max_clics: int):
    """Clique sur 'Montrer plus' / 'Show more' jusqu'à max_clics fois."""
    xpaths = [
        '//button[contains(., "Montrer plus")]',
        '//button[contains(., "Show more")]',
        '//a[contains(@class,"event__more")]',
        '//div[contains(@class,"showMore")]',
    ]
    clicked = 0
    for _ in range(max_clics):
        found = False
        for xp in xpaths:
            try:
                btn = WebDriverWait(driver, 3).until(
                    EC.presence_of_element_located((By.XPATH, xp))
                )
                driver.execute_script(
                    "arguments[0].scrollIntoView({block:'center'});", btn
                )
                time.sleep(0.5)
                driver.execute_script("arguments[0].click();", btn)
                clicked += 1
                time.sleep(DELAY_CLIC)
                found = True
                break
            except Exception:
                continue
        if not found:
            break   # plus de bouton dispo
    return clicked

# ── Scraper par équipe ────────────────────────────────────────────────────────

def scrape_team(driver, team_name: str, url: str) -> list[dict]:
    """
    Scrape jusqu'à N_MATCHES résultats pour une équipe.
    Retourne une liste de dicts avec toutes les colonnes du pipeline.
    """
    print(f"  → {team_name}")
    rows = []

    try:
        driver.get(url)
        time.sleep(DELAY_PAGE)

        accept_cookies(driver)
        n_clics = click_show_more(driver, MAX_CLICS)
        print(f"     {n_clics} clics 'Montrer plus'")

        soup = BeautifulSoup(driver.page_source, "html.parser")
        matchs = soup.find_all(
            "div", class_=lambda x: x and "event__match" in x
        )
        print(f"     {len(matchs)} éléments match trouvés")

        for match in matchs[:N_MATCHES]:

            # ── Résultat via badge de forme ──────────────────────────────
            # (badge présent sur la page résultats Flashscore)
            badge = match.find(
                "button",
                class_=lambda x: x and "formIcon__lastMatches" in x,
            )
            badge_result = badge.text.strip() if badge else None
            # V=win, N=draw, D=loss (labels FR de Flashscore)
            result_from_badge = {"V": "W", "N": "D", "D": "L"}.get(
                badge_result, None
            )

            # ── Équipes & score ──────────────────────────────────────────
            try:
                elem_dom = match.find(
                    class_=lambda x: x and "event__homeParticipant" in x
                )
                elem_ext = match.find(
                    class_=lambda x: x and "event__awayParticipant" in x
                )
                elem_score_dom = match.find(
                    class_=lambda x: x and "event__score--home" in x
                )
                elem_score_ext = match.find(
                    class_=lambda x: x and "event__score--away" in x
                )

                if not all([elem_dom, elem_ext, elem_score_dom, elem_score_ext]):
                    continue

                home_team = elem_dom.text.strip()
                away_team = elem_ext.text.strip()
                score_dom = parse_score(elem_score_dom.text)
                score_ext = parse_score(elem_score_ext.text)

                if score_dom is None or score_ext is None:
                    continue

                # L'équipe cible est-elle à domicile ?
                is_home = team_name.lower() in home_team.lower()

                if is_home:
                    goals_for     = score_dom
                    goals_against = score_ext
                else:
                    goals_for     = score_ext
                    goals_against = score_dom

                # Résultat final : badge en priorité, score en fallback
                result = result_from_badge or result_label(goals_for, goals_against)

            except Exception:
                continue

            # ── Date ─────────────────────────────────────────────────────
            date_el = match.find(
                class_=lambda x: x and "event__time" in x
            )
            date_str = date_el.text.strip() if date_el else ""

            # ── Compétition ───────────────────────────────────────────────
            # On remonte au parent pour chercher le header de compétition
            comp = "Unknown"
            try:
                parent = match.find_parent(
                    "div", class_=lambda x: x and "event__round" in x
                )
                if parent:
                    comp = parent.text.strip()
            except Exception:
                pass

            rows.append({
                "team":          team_name,
                "date":          date_str,
                "home_team":     home_team,
                "away_team":     away_team,
                "home_goals":    score_dom,
                "away_goals":    score_ext,
                "is_home":       is_home,
                "goals_for":     goals_for,
                "goals_against": goals_against,
                "result":        result,
                "competition":   comp,
            })

    except Exception as e:
        print(f"  ⚠ Erreur {team_name} : {e}")

    return rows

# ── Pipeline principal ────────────────────────────────────────────────────────

def run_scraper(df: pd.DataFrame | None = None) -> pd.DataFrame:
    """
    Lance le scraping pour toutes les équipes.

    Args:
        df : DataFrame avec colonnes Team_name & links_flashscore.
             Si None, charge INPUT_CSV automatiquement.

    Returns:
        DataFrame des matchs bruts (sauvegardé dans OUTPUT_CSV).
    """
    if df is None:
        df = pd.read_csv(INPUT_CSV)

    all_matches: list[dict] = []
    driver = make_driver()

    try:
        for idx, row in df.iterrows():
            team = row["Team_name"]
            url  = row.get("links_flashscore", "")

            if pd.isna(url) or not str(url).startswith("http"):
                print(f"  ⚠ URL manquante pour {team}, skip")
                continue

            matches = scrape_team(driver, team, str(url))
            all_matches.extend(matches)
            print(f"     ✓ {len(matches)} matchs récupérés\n")

            time.sleep(DELAY_BTW_TEAMS)

    finally:
        driver.quit()   # toujours fermer même si erreur

    results_df = pd.DataFrame(all_matches)
    OUTPUT_CSV.parent.mkdir(parents=True, exist_ok=True)
    results_df.to_csv(OUTPUT_CSV, index=False)
    print(f"\n✅ Scraping terminé → {len(results_df)} matchs au total")
    print(f"   Sauvegardé : {OUTPUT_CSV}")
    return results_df


# ── Point d'entrée ────────────────────────────────────────────────────────────

if __name__ == "__main__":
    run_scraper()
    
```

```python
"""
ÉTAPE 2 — Feature Engineering
Fusionne les matchs scrapés + les stats CSV pour créer les features d'entraînement.
Sortie : data/features_train.csv

Chaque ligne = un match vu du côté d'une équipe (home ou away).
"""

import re
import pandas as pd
import numpy as np
from pathlib import Path

# ── Config ───────────────────────────────────────────────────────────────────
MATCHES_CSV = Path("matches_raw.csv")
TEAMS_CSV   = Path("stats_equipe.csv")
OUTPUT_CSV  = Path("features_train.csv")
FORM_WINDOW = 5   # fenêtre glissante pour la forme récente

# ── Parsing de la valeur marchande ──────────────────────────────────────────

def parse_value(raw) -> float:
    if not raw or (isinstance(raw, float) and np.isnan(raw)):
        return 0.0
    raw = str(raw).strip().lower()
    raw = re.sub(r'[€\s]', '', raw)
    if 'mio' in raw:
        num = re.sub(r'mio\.?', '', raw).replace(',', '.')
        return float(num) * 1_000_000
    if 'k' in raw:
        num = re.sub(r'k', '', raw).replace(',', '.')
        return float(num) * 1_000
    try:
        return float(raw.replace(',', '.'))
    except ValueError:
        return 0.0

# ── Chargement des données ───────────────────────────────────────────────────

def load_teams(path: Path) -> pd.DataFrame:
    df = pd.read_csv(path)
    df["players_value_eur"] = df["players_value"].apply(parse_value)
    df["win_rate"]           = pd.to_numeric(df["win_rate"],           errors="coerce").fillna(0) / 100
    df["world_cup_win_rate"] = pd.to_numeric(df["world_cup_win_rate"], errors="coerce").fillna(0) / 100
    df["Fifa_ranking"]       = pd.to_numeric(df["Fifa_ranking"],       errors="coerce").fillna(99)
    return df.set_index("Team_name")

def load_matches(path: Path) -> pd.DataFrame:
    df = pd.read_csv(path)
    df["home_goals"] = pd.to_numeric(df["home_goals"], errors="coerce")
    df["away_goals"] = pd.to_numeric(df["away_goals"], errors="coerce")
    df = df.dropna(subset=["home_goals", "away_goals", "result"])
    df["result_num"] = df["result"].map({"W": 2, "D": 1, "L": 0})
    # Buts marqués / encaissés du point de vue de l'équipe cible
    df["goals_for"]     = np.where(df["is_home"], df["home_goals"], df["away_goals"])
    df["goals_against"] = np.where(df["is_home"], df["away_goals"], df["home_goals"])
    return df

# ── Features par équipe (forme récente) ─────────────────────────────────────

def rolling_features(df_team: pd.DataFrame) -> pd.Series:
    """Calcule les features de forme sur les FORM_WINDOW derniers matchs."""
    last = df_team.tail(FORM_WINDOW)
    wins   = (last["result"] == "W").sum()
    draws  = (last["result"] == "D").sum()
    losses = (last["result"] == "L").sum()
    return pd.Series({
        f"form_wins_{FORM_WINDOW}":     wins,
        f"form_draws_{FORM_WINDOW}":    draws,
        f"form_losses_{FORM_WINDOW}":   losses,
        f"form_pts_{FORM_WINDOW}":      wins * 3 + draws,
        f"form_gf_{FORM_WINDOW}":       last["goals_for"].sum(),
        f"form_ga_{FORM_WINDOW}":       last["goals_against"].sum(),
        f"form_gd_{FORM_WINDOW}":       last["goals_for"].sum() - last["goals_against"].sum(),
        f"form_home_wr_{FORM_WINDOW}":  (last[last["is_home"]]["result"] == "W").mean() if last["is_home"].any() else 0,
        f"form_away_wr_{FORM_WINDOW}":  (last[~last["is_home"]]["result"] == "W").mean() if (~last["is_home"]).any() else 0,
    })

# ── Construction des features de match ───────────────────────────────────────

def build_match_features(matches: pd.DataFrame, teams: pd.DataFrame) -> pd.DataFrame:
    """
    Pour chaque match, construit un vecteur de features en combinant :
    - Stats CSV de l'équipe (win_rate, goals, valeur, ranking FIFA)
    - Forme récente (glissante sur les matchs précédents)
    - Avantage domicile
    - Diff de classement FIFA entre les deux équipes
    """
    records = []

    # Grouper par équipe pour calculer la forme au fil du temps
    for team_name, grp in matches.groupby("team"):
        grp = grp.sort_values("date").reset_index(drop=True)

        if team_name not in teams.index:
            continue
        team_stats = teams.loc[team_name]

        for i, match in grp.iterrows():
            # Forme = matchs AVANT ce match (pas de fuite de données)
            past = grp.iloc[:i]
            form = rolling_features(past) if len(past) >= 1 else rolling_features(grp.iloc[:1])

            opponent = match["away_team"] if match["is_home"] else match["home_team"]
            opp_stats = teams.loc[opponent] if opponent in teams.index else None

            rec = {
                # Identifiants
                "team":       team_name,
                "opponent":   opponent,
                "date":       match["date"],
                "is_home":    int(match["is_home"]),
                # Target
                "result":     match["result"],        # W/D/L (classification)
                "goals_for":  match["goals_for"],     # buts marqués (régression)
                "goals_against": match["goals_against"],
                # Stats globales équipe
                "win_rate":          team_stats["win_rate"],
                "world_cup_win_rate":team_stats["world_cup_win_rate"],
                "avg_goals_scored":  team_stats["avg_goals_scored"],
                "avg_goals_conceded":team_stats["avg_goals_conceded"],
                "avg_goal_diff":     team_stats["avg_goal_difference"],
                "players_value":     team_stats["players_value_eur"],
                "fifa_ranking":      team_stats["Fifa_ranking"],
            }

            # Stats adversaire
            if opp_stats is not None:
                rec["opp_win_rate"]      = opp_stats["win_rate"]
                rec["opp_avg_gs"]        = opp_stats["avg_goals_scored"]
                rec["opp_avg_gc"]        = opp_stats["avg_goals_conceded"]
                rec["opp_players_value"] = opp_stats["players_value_eur"]
                rec["opp_fifa_ranking"]  = opp_stats["Fifa_ranking"]
                rec["ranking_diff"]      = team_stats["Fifa_ranking"] - opp_stats["Fifa_ranking"]
                rec["value_ratio"]       = (
                    team_stats["players_value_eur"] / opp_stats["players_value_eur"]
                    if opp_stats["players_value_eur"] > 0 else 1.0
                )
            else:
                rec["opp_win_rate"]      = 0.5
                rec["opp_avg_gs"]        = 0.95
                rec["opp_avg_gc"]        = 1.43
                rec["opp_players_value"] = team_stats["players_value_eur"]/2
                rec["opp_fifa_ranking"]  = 75
                rec["ranking_diff"]      = 0
                rec["value_ratio"]       = 1.0

            # Forme récente
            rec.update(form.to_dict())
            records.append(rec)

    return pd.DataFrame(records)


# ── Main ─────────────────────────────────────────────────────────────────────

def main():
    print("📂 Chargement des données...")
    teams   = load_teams(TEAMS_CSV)
    matches = load_matches(MATCHES_CSV)
    print(f"   {len(teams)} équipes | {len(matches)} matchs bruts")

    print("⚙️  Construction des features...")
    features = build_match_features(matches, teams)
    print(f"   {len(features)} lignes × {len(features.columns)} colonnes")

    OUTPUT_CSV.parent.mkdir(exist_ok=True)
    features.to_csv(OUTPUT_CSV, index=False)
    print(f"✅ Features sauvegardées → {OUTPUT_CSV}")

    # Aperçu
    print("\n── Aperçu des colonnes ──")
    print(features.dtypes) 
    print("\n── Distribution des résultats ──")
    print(features["result"].value_counts())

    return features


if __name__ == "__main__":
    main()
```
